# SIMPLE RATE OF RETURN


```python
#P1-P0/P0 OR P1/P0 - 1
```

```python
df['simple return']=(df['Adj Close']/df['Adj Close'].shift(1))-1
simple = (df['simple return'])
print(simple)
```

```python
simple.plot(figsize=(8,5))
```

```python
avg_return_rate= simple.mean()
print(avg_return_rate)
```

```python
avgrr_annual=avg_return_rate*250
print(avgrr_annual)
```

```python
print(str(round(avgrr_annual,5)*100)+'%')
```

```python
#onestep for annual rate of return in % 
df['simple return']=(df['Adj Close']/df['Adj Close'].shift(1))-1
simple = (df['simple return'])
avg_return_rate= simple.mean()
avgrr_annual=avg_return_rate*250
print(str(round(avgrr_annual,5)*100)+'%')
```

# log rate of return

```python
#log return
df['log return']=np.log(df['Adj Close']/df['Adj Close'].shift(1))
log=(df['log return'])
avg_log_daily=log.mean()
avglogrr_annual=avg_log_daily*250
print(str(round(avglogrr_annual,5)*100)+'%')
```

# simple return

```python
returns=(portfolio/portfolio.shift(1))-1
returns.head()
```

```python
weights=np.array([.25,.25,.25,.25])
```

```python
np.dot(returns,weights)
```

```python
annualreturn= returns.mean()
annualreturn
```

```python
np.dot(annualreturn,weights)
```

```python
portfolio_1= str(round(np.dot(annualreturn,weights),5)*100)+'%'
print(portfolio_1)
```

```python
weights2= np.array([.35,.40,.10,.15])
```

```python
np.dot(annualreturn,weights2)
```

```python
portfolio_2= str(round(np.dot(annualreturn,weights2),5)*100)+'%'
print(portfolio_2)
```

```python
#onestep

returns=(portfolio/portfolio.shift(1))-1
annualreturn= returns.mean()
weights=np.array([.25,.25,.25,.25])
np.dot(annualreturn,weights)
portfolio_1= str(round(np.dot(annualreturn,weights),6)*100)+'%'
print(portfolio_1)
```
