```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
cf=[-200,20,50,70,100,50]
```

```python
x_axis=list(range(1,7))
```

```python
cf_cum=[]
cumul=0
for i in cf:
    cumul+=i
    cf_cum.append(cumul)
cf_cum
```

```python
plt.plot(x_axis,cf)
```

```python
x=list(range(6))
```

```python
plt.plot(x,cf_cum)
```

known as hockeys stick curve

```python
plt.scatter(x,cf)
```

```python
plt.scatter(x,cf_cum)
```

```python
plt.figure(figsize=(12,8))
plt.plot(x,cf_cum, color='red',linestyle='-',marker='o')
#plt.grid()
plt.hlines(y=0,xmin=x[0],xmax=x[-1],linestyle='dashed',color='blue',label='default')
plt.title('Cumulative Cashflow for Project A',fontsize=15)
plt.xlabel('Time in Years')
plt.ylabel('C F',fontsize=12)
plt.yticks(range(-200,101,25))
plt.xticks(x,labels=['t0','t1','t2','t3','t4','t5'])
plt.annotate('Break Even Point',xy=(3.6,0),xytext=(2.5,25),arrowprops={'color':'black'},fontsize=15)
plt.legend(loc='best',fontsize=15)
```

# PLOTTING NPV AND IRR

```python
cf=[-200,20,50,70,100,50]
r=0.06
f=1+r
```

```python
npv=0
for i in range(len(cf)):
    npv+=cf[i]/f**i
print(npv)
    
```

```python
l=list(range(0,151))
```

```python
rs=[]
for i in l:
    rs.append(i/1000)
```

```python
rs
```

```python
len(rs)
```

```python
npvs=[]
for i in rs:
    npv=0
    for j in range(len(cf)):
        npv+=cf[j]/(1+i)**j
    npvs.append(npv)
```

```python
npvs
```

```python
plt.figure(figsize = (12, 8))
plt.scatter(x = r, y = npv, s = 100, c = "blue", marker = "o", label = "NPV @ r")
plt.plot(rs, npvs, color = "red", linewidth = 2, linestyle = "-", label = "NPV(r)")
#plt.grid()
plt.hlines(y = 0, xmin = rs[0], xmax = rs[-1], linestyles= "dashed", color = "blue", label = "NPV = 0")
plt.title("NPV & IRR", fontsize = 15)
plt.xlabel("r", fontsize = 12)
plt.ylabel("NPV", fontsize = 12)
plt.annotate("IRR", xy = (0.119, 0), xytext = (0.135, 20), arrowprops={'color':'black'}, fontsize = 15)
plt.legend(loc = "best", fontsize = 15)
plt.show()
```
