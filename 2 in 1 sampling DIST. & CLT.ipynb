# sampling with np.choice

```python
import numpy as np
```

```python
mu=100
sigma=2
size=1000000
```

```python
np.random.seed(123)
pop=np.random.normal(loc=mu,scale=sigma,size=size)
```

```python
pop
```

```python
pop.mean()
```

```python
pop.std()
```

```python
sample_size=50
```

```python
np.random.seed(123)
sample=np.random.choice(pop,sample_size,replace=False)
```

```python
sample
```

```python
sample.size
```

```python
sample.mean()
```

```python
sample.std(ddof=1)
```

# SAMPLING ERROR

```python
sample.mean()-pop.mean()
```

# SAMPLING DISTRIBUTION

```python
import matplotlib.pyplot as plt
```

```python
pop.size
```

```python
sample_size=10
samples=10000
```

```python
np.random.seed(123)
samples_mean_10=[]
for i in range(samples):
    samples_mean_10.append(np.random.choice(pop,sample_size,replace=False).mean())
```

```python
len(samples_mean_10)
```

```python
sample_size=50
```

```python
np.random.seed(123)
samples_mean_50=[]
for i in range(samples):
    samples_mean_50.append(np.random.choice(pop,sample_size,replace=False).mean())
```

```python
samples_mean_50
```

```python
len(samples_mean_50)
```

```python
plt.figure(figsize=(12,8))
plt.hist(samples_mean_10,bins=200,alpha=0.5,label='Sample Size 10')
plt.hist(samples_mean_50,bins=200,alpha=0.5,label='Sample Size 50')
plt.title('Sample Distribution of Mean')
plt.xlabel('Sample means')
plt.ylabel('Absolute Frequency')
plt.legend(fontsize=14)
```

# CENTRAL LIMIT THEOREM

ANY DISTRIBUTION(UNIFORM OR NORMAL) APPROACHES NORMAL DISTRIBUTION AS WE INCREASE SAMPLE SIZE(ABOVE 30)

```python
import scipy.stats as stats
```

```python
pop
```

```python
pop.mean()
```

```python
pop.std()
```

```python
sample_size=2
samples=10000
```

```python
stderror=pop.std()/np.sqrt(sample_size)
stderror
```

```python
sample_mean=pop.mean()
```

```python
np.random.seed(123)
samples_mean_2=[]
for i in range(samples):
    samples_mean_2.append(np.random.choice(pop,sample_size,replace=False).mean())
```

```python
len(samples_mean_2)
```

```python
np.std(samples_mean_2)
```

```python
np.mean(samples_mean_2)
```

```python
x=np.linspace(96,104,1000)
y=stats.norm.pdf(x,loc=sample_mean,scale=stderror)
```

```python
plt.figure(figsize=(12,8))
plt.hist(samples_mean_2,bins=100,alpha=0.5,label='Sample Size 2',density=True)
plt.plot(x,y,color='red',label='Normal Distribution')
plt.title('Sample Distribution of Mean')
plt.xlabel('Sample means')
plt.ylabel('PDF')
plt.legend(fontsize=14)
```

# CENTRAL LIMIT THEOREM CONTINOUS UNIFORM

```python
np.random.seed(123)
pop=np.random.uniform(0,10,100000)
```

```python
pop
```

```python
plt.figure(figsize=(12,8))
plt.hist(pop,bins=100,density=True)
plt.ylabel('pdf')
```

```python
pop.mean()
```

```python
pop.std()
```

```python
sample_size=2
samples=100000
```

```python
sample_means=[]
for i in range(samples):
    sample_means.append(np.random.choice(pop,sample_size,replace=False).mean())
```

```python
b=np.std(sample_means)
```

```python
a=np.mean(sample_means)
```

```python
x=np.linspace(0,10,1000)

y=stats.norm.pdf(x,loc=a,scale=b)
```

```python
plt.figure(figsize=(12,8))
plt.hist(sample_means,bins=100,alpha=0.5,label=f'Sample Size: {sample_size}',density=True)
plt.plot(x,y,color='red',label='Normal Distribution')
plt.legend(fontsize=14)
```
